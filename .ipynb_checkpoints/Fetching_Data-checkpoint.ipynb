```python
import requests
import pandas as pd

# Set the API endpoint and parameters
count=0
for sym in list_of_symbol:
    if count !=100:
        try:
            endpoint = 'https://min-api.cryptocompare.com/data/v2/histoday'
            params = {
                'fsym': '365',         # Cryptocurrency symbol
                'tsym': 'USD',         # Quote currency symbol
                'limit': 100,          # Number of data points to retrieve
            #     'aggregate': 1,        # Time period to aggregate the data over (in days)
                'toTs': pd.Timestamp.now().floor('D').timestamp()  # End timestamp (today)
            }

            # Make the API request
            response = requests.get(endpoint, params=params)

            # Extract the data from the response
            data = response.json()['Data']['Data']

            # Convert the data to a pandas DataFrame
            df = pd.DataFrame(data)
            df = df[['time', 'open', 'high', 'low', 'close']]
            df['time'] = pd.to_datetime(df['time'], unit='s')
            df.set_index('time', inplace=True)
            df.sort_index(inplace=True)
            df.to_csv(f"Symbol_{sym}.csv")
            print(df)
            count+=1
            print(f"Done for {sym}")
        except Exception as e:
            print(f"Error in {sym}")
```

```python
import requests
import pandas as pd

df = pd.read_csv('sym.csv')
x = df['Symbol'].tolist()
for sym in x:
    endpoint = 'https://min-api.cryptocompare.com/data/v2/histoday'
    params = {
        'fsym': sym,         # Cryptocurrency symbol
        'tsym': 'USD',         # Quote currency symbol
        'limit': 365,          # Number of data points to retrieve
    #     'aggregate': 1,        # Time period to aggregate the data over (in days)
        'toTs': pd.Timestamp.now().floor('D').timestamp()  # End timestamp (today)
        }

    # Make the API request
    response = requests.get(endpoint, params=params)
    # Extract the data from the response
    data = response.json()['Data']['Data']
    # Convert the data to a pandas DataFrame
    df = pd.DataFrame(data)
    df = df[['time', 'open', 'high', 'low', 'close']]
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.set_index('time', inplace=True)
    df.sort_index(inplace=True)
    df.to_csv(f"Symbol_{sym}.csv")
```
